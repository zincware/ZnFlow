```python
import znflow
import dataclasses
```

```python
@znflow.nodify
def summmation(*args):
    return sum(args)
```

```python
@dataclasses.dataclass
class AddNumbers(znflow.Node):
    a: int
    b: int
    results: int = -1
    
    def run(self):
        self.results = self.a + self.b
```

```python
a = summmation(1, 2, 3)
b = summmation(4, 5, 6)
c = summmation(a, b)

print(c)
```

```python
with znflow.DiGraph() as graph:
    print(AddNumbers._graph_)
    a = summmation(1, 2, 3)
    b = summmation(4, 5, 6)
    c = summmation(a, b)
    d = AddNumbers(7, 8)
    e = summmation(c, d.results)
    
print(AddNumbers._graph_)
```

```python
@dataclasses.dataclass
class Node(znflow.Node):
    inputs: float
    outputs: float = None
    
    def run(self):
        self.outputs = self.inputs * 2

@dataclasses.dataclass
class SumNodes(znflow.Node):
    inputs: float
    outputs: float = None

    def run(self):
        self.outputs = sum(self.inputs)
```

```python
import random
k = 3
j = 3
i = 3

with znflow.DiGraph() as graph:
    kdx_nodes = []
    for kdx in range(k):
        jdx_nodes = []
        for jdx in range(j):
            idx_nodes = []
            for idx in range(i):
                idx_nodes.append(Node(inputs=random.random()))
            jdx_nodes.append(SumNodes(inputs=[x.outputs for x in idx_nodes]))
        kdx_nodes.append(SumNodes(inputs=[x.outputs for x in jdx_nodes]))
            
    end_node = SumNodes(inputs=[x.outputs for x in kdx_nodes])
```

```python
znflow.draw(graph)
```
