# Functions and Graphs

```python
import znflow
```

```python
@znflow.nodify
def add(*args):
    return sum(args)

@znflow.nodify
def multiply(a, b):
    return a * b

@znflow.nodify
def divide(a, b):
    print("Computing")
    return a / b
```

_eager_ mode

```python
n1 = add(1, 2, 3)
n2 = add(10, 20, 30)
n3 = add(n1, n2)
n4 = multiply(n1, n3)
n5 = divide(n4, n1)
```

```python
n5
```

_graph_ mode

```python
with znflow.DiGraph() as graph:
    n1 = add(1, 2, 3)
    n2 = add(10, 20, 30)
    n3 = add(n1, n2)
    n4 = multiply(n1, n3)
    n5 = divide(n4, n1)
```

```python
znflow.draw(graph, log=False)
```

```python
graph.run()
```

```python
n5.result
```
